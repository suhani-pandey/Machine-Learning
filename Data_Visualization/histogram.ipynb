# Created a histogram with random data

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=20, edgecolor='black')

# Add title and labels
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

# histogram with multiple data sets

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  # Shifted data

# Create histogram
# alpha is the transparency of the bars
plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1', edgecolor='black')
plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2', edgecolor='black')

# Add title and labels
plt.title('Histogram of Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

# histogram with cumulative frequency

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create histogram with cumulative frequency
plt.hist(data, bins=30, cumulative=True, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')

# Show the plot
plt.show()
```

# horizontal histogram
It is used to display the distribution of data when the labels are long or there are many categories.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create horizontal histogram
plt.hist(data, bins=30, orientation='horizontal', color='salmon', edgecolor='black')

# Add title and labels
plt.title('Horizontal Histogram')
plt.xlabel('Frequency')
plt.ylabel('Value')

# Show the plot
plt.show()
```

# histogram with density plots
It is useful to overlay a density plot on top of a histogram to better understand the distribution of the data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create histogram with density plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')

# Overlay with density plot
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

# Add title and labels
plt.title('Histogram with Density Plot')
plt.xlabel('Value')
plt.ylabel('Density')

# Show the plot
plt.show()
```
