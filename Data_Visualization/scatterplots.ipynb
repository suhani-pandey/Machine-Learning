# Simple scatterplots example

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 4, 5, 4, 7]  # Y-axis values

# Create scatterplot
plt.scatter(x, y, color='blue')

# Add title and labels
plt.title('Simple Scatterplot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

# Scatterplots with different colors
This example shows how to create a scatterplot with different colors for each point. This is used to highlight different groups of data points.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # Random colors for each point

# Create scatterplot
plt.scatter(x, y, c=colors, cmap='viridis', edgecolor='black')

# Add colorbar
plt.colorbar(label='Color intensity')

# Add title and labels
plt.title('Scatterplot with Colored Points')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```

Use Case: Visualizing different regions on a map based on population density or temperature.

Scenario: Suppose you are analyzing cities based on their population and economic output. You can use the X-axis for population and the Y-axis for economic output. Then, you can color each point based on the average income in that city.


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: population (millions), economic output (GDP), average income ($)
population = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.0])
economic_output = np.array([300, 400, 600, 800, 850, 950])
average_income = np.array([20000, 25000, 18000, 22000, 30000, 40000])  # Used for coloring

# Create scatterplot
plt.scatter(population, economic_output, c=average_income, cmap='viridis', edgecolor='black')

# Add colorbar
plt.colorbar(label='Average Income')

# Add title and labels
plt.title('Population vs Economic Output (Colored by Income)')
plt.xlabel('Population (Millions)')
plt.ylabel('Economic Output (GDP in Millions)')

# Show the plot
plt.show()
```

# Scatterplots with size variation

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 1000 * np.random.rand(50)  # Random sizes for each point

# Create scatterplot
plt.scatter(x, y, s=sizes, color='blue', alpha=0.5, edgecolor='black')

# Add title and labels
plt.title('Scatterplot with Size Variation')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```

Use Case: Market analysis to visualize customer spending, transaction count, and customer age.

Scenario: Imagine you are working in retail and want to analyze customer behavior. The X-axis could be customer age, the Y-axis could be the number of transactions they make per year, and the size of each point could represent the total spending by each customer.


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: customer age, number of transactions, total spending ($)
age = np.array([25, 34, 45, 53, 22, 30, 40, 60])
transactions = np.array([15, 18, 20, 12, 25, 30, 10, 8])
spending = np.array([2000, 3000, 4000, 2500, 1500, 5000, 1000, 800])  # Used for point size

# Create scatterplot
plt.scatter(age, transactions, s=spending, color='blue', alpha=0.5, edgecolor='black')

# Add title and labels
plt.title('Customer Age vs Transactions (Point Size = Spending)')
plt.xlabel('Customer Age')
plt.ylabel('Number of Transactions')

# Show the plot
plt.show()
```

# Scatterplots with labels on Data points

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
labels = ['A', 'B', 'C', 'D', 'E']  # Labels for each point

# Create scatterplot
plt.scatter(x, y, color='red')

# Add labels to each point
for i in range(len(x)):
    plt.text(x[i], y[i], labels[i], fontsize=12, ha='right')

# Add title and labels
plt.title('Scatterplot with Labels')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```

Use Case: Labeling outliers or important data points.

Scenario: You are analyzing the performance of different companies based on their revenue and profits. Each point represents a company, and the labels display the companys name.


```python
import matplotlib.pyplot as plt

# Sample data: revenue (millions), profit (millions), company names
revenue = [100, 200, 150, 300, 400]
profit = [10, 40, 25, 60, 80]
companies = ['Company A', 'Company C', 'Company D', 'Company E', 'Company B']

# Create scatterplot
plt.scatter(revenue, profit, color='red')

# Add labels to each point
for i in range(len(revenue)):
    plt.text(revenue[i], profit[i], companies[i], fontsize=12, ha='right')

# Add title and labels
plt.title('Revenue vs Profit (Labeled with Company Names)')
plt.xlabel('Revenue (Millions)')
plt.ylabel('Profit (Millions)')

# Show the plot
plt.show()
```

# Scatterplot with regression line

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + np.random.rand(50) * 0.5  # Linear relationship with noise

# Create scatterplot
plt.scatter(x, y, color='blue', edgecolor='black')

# Add a trend line (linear regression)
m, b = np.polyfit(x, y, 1)  # Fit a line to the data
plt.plot(x, m * x + b, color='red')  # Plot the line

# Add title and labels
plt.title('Scatterplot with Regression Line')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```

Use Case: Predicting future values based on existing data trends.

Scenario: Suppose you are analyzing advertising spend (X-axis) vs. sales (Y-axis) for a company. You want to see if there is a relationship between how much the company spends on ads and how much it makes in sales. Adding a regression line shows the trend.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: advertising spend ($), sales ($)
ad_spend = np.array([100, 200, 300, 400, 500])
sales = np.array([20, 50, 70, 100, 130])

# Create scatterplot
plt.scatter(ad_spend, sales, color='blue', edgecolor='black')

# Add a trend line (linear regression)
m, b = np.polyfit(ad_spend, sales, 1)  # Fit a line to the data
plt.plot(ad_spend, m * ad_spend + b, color='red')  # Plot the line

# Add title and labels
plt.title('Advertising Spend vs Sales (With Regression Line)')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Sales ($)')

# Show the plot
plt.show()
```

# Scatterplot with multiple categories

```python
import matplotlib.pyplot as plt

# Data points for two categories
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
x2 = [1, 2, 3, 4, 5]
y2 = [3, 5, 7, 9, 11]

# Create scatterplot for category 1
plt.scatter(x1, y1, color='blue', label='Category 1')

# Create scatterplot for category 2
plt.scatter(x2, y2, color='green', label='Category 2')

# Add title and labels
plt.title('Scatterplot with Multiple Categories')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Add legend to differentiate the categories
plt.legend()

# Show the plot
plt.show()
```

Use Case: Comparing performance across different groups or categories.

Scenario: In an educational setting, you might want to compare the scores of two different classes on a math test. Class 1's scores are plotted in blue, and Class 2's scores are plotted in green.

```python
import matplotlib.pyplot as plt

# Sample data: test scores for two classes
class1_scores = [75, 80, 85, 90, 95]
class2_scores = [70, 78, 82, 88, 92]
students = [1, 2, 3, 4, 5]  # Student identifiers

# Create scatterplot for Class 1
plt.scatter(students, class1_scores, color='blue', label='Class 1')

# Create scatterplot for Class 2
plt.scatter(students, class2_scores, color='green', label='Class 2')

# Add title and labels
plt.title('Comparison of Test Scores Between Two Classes')
plt.xlabel('Student')
plt.ylabel('Test Score')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

# Pair plots for multiple scatterplots

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a simple dataset for two patient groups (Group A and Group B)
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 28, 33, 38, 43, 48],
    'Cholesterol': [180, 190, 195, 210, 230, 240, 250, 260, 265, 270, 175, 185, 200, 215, 225],
    'Blood Pressure': [120, 125, 130, 140, 145, 150, 155, 160, 165, 170, 118, 123, 128, 138, 142],
    'Group': ['Group A'] * 10 + ['Group B'] * 5  # First 10 entries for Group A, next 5 for Group B
}

# Convert the dictionary into a DataFrame
df = pd.DataFrame(data)

# Create a pair plot to visualize the relationships between the health metrics
sns.pairplot(df, hue='Group', diag_kind='kde')  # diag_kind='kde' shows kernel density plots on the diagonal
plt.show()
```
